```python
import numpy as np
import matplotlib.pyplot as plt

# Load MNIST (from keras just for data)
from tensorflow.keras.datasets import mnist

# Load data
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.reshape(-1, 28*28) / 255.0  # normalize 0-1
x_test = x_test.reshape(-1, 28*28) / 255.0

# One-hot encode labels (for input)
def one_hot(labels, num_classes=10):
    return np.eye(num_classes)[labels]

y_train_onehot = one_hot(y_train)
y_test_onehot = one_hot(y_test)

# --- Neural network generator ---
# input: one-hot (10,)
# output: image (784,)
# structure: 10 -> 128 -> 784

class SimpleGenerator:
    def __init__(self):
        # Xavier initialization
        self.W1 = np.random.randn(10, 128) * np.sqrt(1/10)
        self.b1 = np.zeros((1, 128))
        self.W2 = np.random.randn(128, 784) * np.sqrt(1/128)
        self.b2 = np.zeros((1, 784))

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def forward(self, x):
        self.z1 = x @ self.W1 + self.b1
        self.a1 = np.tanh(self.z1)
        self.z2 = self.a1 @ self.W2 + self.b2
        self.a2 = self.sigmoid(self.z2)  # output is pixel values (0-1)
        return self.a2

    def backward(self, x, y, lr=0.01):
        m = x.shape[0]
        # forward
        out = self.forward(x)
        # loss derivative (MSE)
        dz2 = (out - y) * (out * (1 - out))
        dW2 = self.a1.T @ dz2 / m
        db2 = np.sum(dz2, axis=0, keepdims=True) / m

        dz1 = (dz2 @ self.W2.T) * (1 - np.tanh(self.z1)**2)
        dW1 = x.T @ dz1 / m
        db1 = np.sum(dz1, axis=0, keepdims=True) / m

        # update
        self.W1 -= lr * dW1
        self.b1 -= lr * db1
        self.W2 -= lr * dW2
        self.b2 -= lr * db2

        # return loss
        loss = np.mean((y - out)**2)
        return loss

# --- Train ---
gen = SimpleGenerator()

epochs = 25
batch_size = 128

for epoch in range(epochs):
    perm = np.random.permutation(len(x_train))
    x_train_shuffled = x_train[perm]
    y_train_shuffled = y_train_onehot[perm]

    losses = []
    for i in range(0, len(x_train), batch_size):
        batch_x = x_train_shuffled[i:i+batch_size]
        batch_y = y_train_shuffled[i:i+batch_size]
        loss = gen.backward(batch_y, batch_x, lr=0.1)  # input = label, output = image
        losses.append(loss)

    print(f"Epoch {epoch+1}, Loss: {np.mean(losses):.4f}")

# --- Generate a digit ---
def generate_digit(number):
    onehot = one_hot(np.array([number]))
    img = gen.forward(onehot).reshape(28,28)
    plt.imshow(img, cmap="gray")
    plt.title(f"Generated Digit: {number}")
    plt.axis("off")
    plt.show()

# Example: generate digit 7
generate_digit(2)
```
